```python
data1 = {'Name': ['Jai', 'Hari', 'Gaurav', 'Anuj'], 
         'Age':[27, 24, 22, 32],
         'Address':['Nagpur', 'Kanpur', 'Allahabad','Kannada'],
         'Qualification':['Msc', 'MA', 'ME', 'Phd']}
data2 = {'Name': ['Jaya', 'Harish', 'Anuja', 'Tanuja'], 
         'Age':[17, 14, 12, 42],
         'Address':['Nagpur', 'Kanpur', 'Allahabad','Kannada'],
         'Qualification':['Btech', 'BA', 'BSc', 'BArch']}
df = pd.DataFrame(data1, index=[0,1,2,3])
df1 = pd.DataFrame(data2, index=[4,5,6,7])
print(df, "\n\n", df1)
```

```python
frames = [df, df1]
result = pd.concat(frames)
print(result)
```

```python
df_new = pd.DataFrame(data1, index=[0,1,2,3])
df1 = pd.DataFrame(data2, index=[2,3,4,5])
#print(df, "\n\n", df1)

result2 = pd.concat([df_new, df1], axis=1, join='inner') #by index
print(result2)
result2 = pd.concat([df_new, df1], axis=1, join='outer') #by index
print(result2)
```

```python
#Defining the dictionary
data1 = {'Key':['k0','k1','k2','k3'],
         'Name': ['Jai', 'Hari', 'Gaurav', 'Anuj'],
          'Age':[27, 24, 22, 32]}
data2 = {'Key':['k0','k1','k2','k3'],
         'Address':['Nagpur', 'Kanpur', 'Allahabad','Kannada'],
          'Qualification':['Btech', 'BA', 'BSc', 'BArch']}

#convert the dictionary into dataframe
df = pd.DataFrame(data1)
df1 = pd.DataFrame(data2)

print(df, '\n\n', df1)

res = pd.merge(df, df1, on = 'Key')
res
```
